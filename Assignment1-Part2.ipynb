**Write a function to check whether the number is Palindrome or not.**  
- Hopefully who have did this in c++ in yor CS101

```python
test_num = int(input("Enter your nuber: "))
elements = list()

for i in str(test_num):
    elements.append(int(i))

for i in range(len(elements)):
    if elements[i] != elements[-(i+1)]:
        print("Not a Palindrome")
        check = False
        break
    else:
        check = True

if check:
    print("Palindrome")
```


Create a NumPy array of 10 random integers between 1 and 100 and calculate the mean, minimum, and maximum values


```python
import numpy as np
import random

num = list()
for i in range(10):
    num.append(random.randint(1,100))

arr = np.array(num)
print(arr)
print(f"Mean : {arr.mean()}")
print(f"Min : {arr.min()}")
print(f"Max : {arr.max()}")
```


Create a Pandas DataFrame from a dictionary containing three columns: 'Name', 'Age', and 'City'. Print the DataFrame.
filter the rows where the 'Age' is greater than 30 and print the result.
Add a new column 'Salary' to the DataFrame created in Exercise 1 with values [50000, 60000, 70000]. Print the updated DataFrame
Group the DataFrame by the 'City' column and calculate the average 'Age' for each city. And plot appropriate/required graphs.

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({ 
    "Name": ["John", "Max", "Mike"],
    "Age": [18,45,27],
    "City": ["New York", "Chicago", "New York"],
    })

print(df)

age_Above_30 = df[df["Age"] > 30]

print(age_Above_30)

df.insert(len(df.columns), "Salary", [50000, 60000, 70000])

print(df)

group_city = df.groupby("City")

print(group_city["Age"].mean())

city_list = list()

for i in df["City"]:
    if i not in city_list:
        city_list.append(i)

plt.bar(city_list, group_city["Age"].mean())
plt.title('Avg. Age of Cities')
plt.xlabel('City')
plt.ylabel('Avg. Age')
plt.show()
```
