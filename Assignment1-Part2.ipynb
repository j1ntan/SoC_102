**Write a function to check whether the number is Palindrome or not.**  
- Hopefully who have did this in c++ in yor CS101

```python
test_num = int(input("Enter your nuber: "))
elements = list()

for i in str(test_num):
    elements.append(int(i))

for i in range(len(elements)):
    if elements[i] != elements[-(i+1)]:
        print("Not a Palindrome")
        check = False
        break
    else:
        check = True

if check:
    print("Palindrome")
```


Create a NumPy array of 10 random integers between 1 and 100 and calculate the mean, minimum, and maximum values


```python
import numpy as np
import random

num = list()
for i in range(10):
    num.append(random.randint(1,100))

arr = np.array(num)
print(arr)
print(f"Mean : {arr.mean()}")
print(f"Min : {arr.min()}")
print(f"Max : {arr.max()}")
```
